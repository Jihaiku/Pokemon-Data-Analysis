#### New type of Visualization
I was researching how other people had been doing this project and thanks to Andrew Gele on Kaggle I learned about the swarmplot and I thought it would be helpful to use it here.

```python
pkmn = pd.melt(df_stats, id_vars=['name', 'type1', 'type2', 'is_legendary'], var_name='stat')
pkmn.head()
```

```python
sns.swarmplot(x="stat", y="value", data=pkmn, hue="type1")
```

This is essentially a recreation but with new data of his design and while I think that the swarmplot is very interesting, I personally have a very hard time reading one due to the color scale and I tried to find an alternative where color wasn't the primary tool to discern them but it seems that aside from using shapes (which is an entirely different mess) there wasn't a real way to alter it to work with color blind mode.

### Final Version of swarmplot
So instead I just went through with the cleaner version of the swarmplot. It's not quite the hexagonal shaped visualization I was wanting to do for this update like in the older pokemon games but it's a step closer I think.

```python
plt.figure(figsize=(12,10))
plt.ylim(0, 275)
sns.swarmplot(x="stat", y="value", data=pkmn, hue="type1", split=True, size=7)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.);
```
